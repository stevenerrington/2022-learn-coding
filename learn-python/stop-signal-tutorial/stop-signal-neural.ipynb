# Neural data from the stop-signal dataset

## Import libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plot
```

## Load data

```python
dataDir = "./data/";

# Event times
eventTimes = pd.read_csv(dataDir+ "dar-cmand1DR-20201127-events.csv")
eventInfos = pd.read_csv(dataDir+ "dar-cmand1DR-20201127-info.csv")

# Neural data
spikeTimes = np.load(dataDir + "neural_data/spike_times.npy")*(24414.14/1000);
spikeClusters = np.load(dataDir + "neural_data/spike_clusters.npy");
spikeAmplitudes = np.load(dataDir + "neural_data/amplitudes.npy");
```

## Neural data

```python
# Get a list of the clusters identified through Kilosort (& curated through Phy)
ks_clust_list = np.unique(spikeClusters)
```

### ISI Histogram

```python
clust = 6;
spkTimes_cluster = np.transpose(spikeTimes[spikeClusters == clust])
```

```python
ISI_dist = np.diff(spkTimes_cluster)
```

### Plot clusters

```python
# For each cluster, plot the amplitude over time
for clust in ks_clust_list:
    plot.figure() # Open a new figure
    plot.scatter(spikeTimes[spikeClusters == clust],spikeAmplitudes[spikeClusters == clust])
    plot.xlabel("Time in session (ms)"); plot.ylabel("Amplitude (a.u.)")
```
