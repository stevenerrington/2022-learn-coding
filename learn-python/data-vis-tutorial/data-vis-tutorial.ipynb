# Data visualation in matplotlib

```python
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
```

## Creating a generic standalone plot (i.e. scatter, line)

```python
##################################################################
# Line plot with markers

# Input data
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

# Initiate plot type
plt.plot(x, y, marker = 'o')

# Decorate plot
plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

# Generate plot
plt.show()

##################################################################
# Scatter plot with two sets of data
#      Data is added and defined, before the plot is generated

# Scatter dataset 1
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
plt.scatter(x, y, color = 'hotpink', alpha=0.5)

# Scatter dataset 2
x = np.array([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12])
y = np.array([100,105,84,105,90,99,90,95,94,100,79,112,91,80,85])
plt.scatter(x, y, color = '#88c999', alpha=0.5)

# Generate plot
plt.show()

##################################################################
# Bar plot

x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x, y, color = "#4CAF50")
plt.show()

##################################################################
# Pie chart plot

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels)
plt.show() 

##################################################################
# Histogram

x = np.random.normal(170, 10, 250)

plt.hist(x)
plt.show() 
```

## Creating a multi-subplot figure

```python
# Subplots are defined by plt.subplot(rows,columns,subplot number)
# Similar to matlab

# Subplot 1
x = np.array([0, 1, 2, 3]); y = np.array([3, 8, 1, 10])
plt.subplot(2, 2, 1); plt.plot(x,y)

# Subplot 2
x = np.array([0, 1, 2, 3]); y = np.array([10, 20, 30, 40])
plt.subplot(2, 2, 2); plt.plot(x,y)

# Subplot 3
x = np.array([0, 1, 2, 3]); y = np.array([3, 8, 1, 10])
plt.subplot(2, 2, 3); plt.plot(x,y)

# Subplot 4
x = np.array([0, 1, 2, 3]); y = np.array([10, 20, 30, 40])
plt.subplot(2, 2, 4); plt.plot(x,y)

# Generate plot
plt.show()
```
